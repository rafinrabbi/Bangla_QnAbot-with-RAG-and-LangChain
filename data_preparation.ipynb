## Text extraction
### Attempt 1
NOTE:
We did NOT use the extracted text from the PDF for further processing because the Bangla font appears broken or garbled after extraction.
This is a common issue with PDF text extraction libraries (like PyMuPDF) when handling complex scripts such as Bangla.
The extracted text may lose formatting, contain missing or incorrect characters, or become unreadable.
For Bangla or similar languages, consider using OCR tools (e.g., Tesseract with Bangla language pack) or obtain the text from a reliable digital source.


```python
import fitz  # PyMuPDF


def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text("text")
    return text

# Provide your PDF path here
pdf_path = './data/HSC26-Bangla1st-Paper.pdf'
text = extract_text_from_pdf(pdf_path)
print(text)

```

## Text extraction
### Attempt 2

Note: For OCR processing, we first save each page of the PDF as an image. This is necessary because OCR tools like Tesseract work on image files, not directly on PDFs. The following code converts each PDF page into a PNG image for further OCR-based text extraction.


```python


# Import the required libraries
from pdf2image import convert_from_path
import os

# I need it for my ubuntu machine but works fine in Google Colab
# Set the TESSDATA_PREFIX environment variable to the path where Tesseract's language data
os.environ['TESSDATA_PREFIX'] = '/usr/share/tesseract-ocr/4.00/tessdata/'


# Define the folder path to save images (change this to your desired folder)
folder_path = './data/pdf2img'

# Create the folder if it doesn't exist
os.makedirs(folder_path, exist_ok=True)

# Convert PDF to images
images = convert_from_path('./data/HSC26-Bangla1st-Paper.pdf')

# Save each image to the folder
for i, image in enumerate(images):
    image_path = os.path.join(folder_path, f"page_{i + 1}.png")  # Save as PNG
    image.save(image_path, 'PNG')

print(f"Images have been saved to {folder_path}")
```

Note:
While using pytesseract for OCR, it was observed that not all text was accurately extracted from the PDF. For example, the MCQ questions on page 2 were missed, and many Bangla conjunct characters appeared broken or unreadable. This highlights the limitation of OCR for complex scripts and structured content in Bangla PDFs. Manual review and correction may be necessary for complete and accurate extraction.

```python
import pytesseract
from PIL import Image
import os

# Path to the folder where images are saved
folder_path = './data/pdf2img'

# Get sorted list of image files
image_files = sorted([f for f in os.listdir(folder_path) if f.endswith('.png')])

# Loop through images and apply OCR
for i, image_file in enumerate(image_files):
    image_path = os.path.join(folder_path, image_file)
    image = Image.open(image_path)
    # Apply OCR for Bangla text
    text = pytesseract.image_to_string(image, lang='ben')  # 'ben' is the language code for Bangla
    print(f"Page {i+1}:\n{text}")

# Open a text file to save the extracted text
with open('./data/extracted_text/output_text_pytesseract.txt', 'w', encoding='utf-8') as file:
    for i, image_file in enumerate(image_files):
        image_path = os.path.join(folder_path, image_file)
        image = Image.open(image_path)
        text = pytesseract.image_to_string(image, lang='ben')
        file.write(f"Page {i+1}:\n{text}\n\n")

print("OCR text has been saved to 'output_text.txt'.")
```

## Text extraction with Gemini OCR
### Attempt 3

**Note:**  
After some research and experimentation, I found that Gemini OCR works very well for text extraction from images. It is almost flawless and did not miss any text blocks in my tests, even with complex Bangla scripts and layouts. This makes Gemini OCR a strong choice for high-quality and accurate text extraction.

```python
# !pip install google-generativeai
!pip install python-dotenv
```

```python

import os
import base64
import google.generativeai as genai
from PIL import Image
import io
from dotenv import load_dotenv

# Load environment variables (make sure .env file exists or set the key manually)
load_dotenv()
# Or set directly: os.environ['GEMINI_API_KEY'] = 'your_actual_api_key'

def process_image(image_path):
    """Process a single image with Gemini API and return description"""
    try:
        # Initialize Gemini model
        genai.configure(api_key=os.environ.get("GEMINI_API_KEY"))
        model = genai.GenerativeModel('gemini-2.0-flash')  # Use a supported model

        # Open and encode image
        with Image.open(image_path) as img:
            img_byte_arr = io.BytesIO()
            img.save(img_byte_arr, format='PNG')
            img_data = img_byte_arr.getvalue()  # Get raw bytes for Gemini

        # Prepare content for Gemini
        contents = [
            {
                "parts": [
                    {"text": "Extract the plain text"},
                    {
                        "inline_data": {
                            "mime_type": "image/png",
                            "data": base64.b64encode(img_data).decode('utf-8')
                        }
                    }
                ]
            }
        ]

        # Generate description
        response = model.generate_content(contents)
        return response.text

    except Exception as e:
        return f"Error processing {image_path}: {str(e)}"


# Define input and output directories (local paths)
input_dir = './data/pdf2img'  # Your input folder path
output_dir = './data/extracted_text_gemini'  # Your output folder path

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Supported image extensions
image_extensions = ('.png', '.jpg', '.jpeg', '.bmp')

# Process each image in the input directory
for filename in os.listdir(input_dir):
    if filename.lower().endswith(image_extensions):
        image_path = os.path.join(input_dir, filename)

        # Process image and get description
        description = process_image(image_path)

        # Save description to output directory
        output_filename = os.path.splitext(filename)[0] + '_description.txt'
        output_path = os.path.join(output_dir, output_filename)

        with open(output_path, 'w', encoding='utf-8') as f:
            f.write(description)

        print(f"Processed {filename} and saved description to {output_filename}")

print("All images processed successfully.")
```

```python

import os

# Local directory containing the Gemini OCR output files
folder_path = './data/extracted_text_gemini'

# Output file path for the merged content
output_file_path = os.path.join(folder_path, "merged_gemini_output.txt")

# List all text files in the folder that match the pattern
text_files = [f for f in os.listdir(folder_path) if f.endswith('_description.txt')]

# Sort files based on the page number (assuming files are named page_1_description.txt, page_2_description.txt, etc.)
text_files.sort(key=lambda x: int(x.split('_')[1]))

# Open the output file for writing
with open(output_file_path, 'w', encoding='utf-8') as output_file:
    # Iterate over each file, read its content, and write to the output file
    for text_file in text_files:
        file_path = os.path.join(folder_path, text_file)
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
            output_file.write(content + '\n\n')  # Add two newlines for separation between pages

print(f"All text files have been successfully combined into {output_file_path}")
```

### Data Cleaning

```python
import re

# Example function to clean and preprocess the extracted text
def clean_text(text):
    # Remove unwanted characters, page numbers, headers, footers, etc.
    text = re.sub(r'\d{1,3}\s?(\.|:|-|\))', '', text)  # Remove page numbers like "10."
    text = re.sub(r'\n+', '\n', text)  # Remove extra newlines
    text = text.strip()  # Remove leading/trailing whitespaces
    return text

# Example for cleaning the text from file
input_path = './data/extracted_text_gemini/merged_gemini_output.txt'
output_path = './data/extracted_text_gemini/cleaned_gemini_output.txt'

with open(input_path, 'r', encoding='utf-8') as file:
    raw_text = file.read()
    cleaned_text = clean_text(raw_text)

# Save the cleaned text to a new file
with open(output_path, 'w', encoding='utf-8') as file:
    file.write(cleaned_text)

# Check the cleaned text
print(cleaned_text[:1500])  # Show the first 500 characters
```
